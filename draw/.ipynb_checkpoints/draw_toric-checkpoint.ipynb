```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
import matplotlib.patches as pat
from matplotlib.collections import LineCollection

from draw_toric import *
```

```python
l1 = 5
l2 = 5
lattice55 = make_lattice(l1,l2)

vertices = lattice55[0]
faces = lattice55[1]
```

```python
erasure_vector_for_qubits  = [0,6,8,10,21,33,38,41,44,49]
Z_errors_on_qubits  = [6,8,21,38,41,44,49]
X_syndromes = [2,4,5,10,15,18,20,21,22,24]
```

```python
# draw codes
fig, ax = plt.subplots(figsize=(l1, l2))

# vertices, Z stab
for i in vertices:
    plt.plot(i[0], i[1],'.',markersize=10, color = "gray")
    
# edges, qubit
lines = []
for e in erasure_vector_for_qubits:
    cdn = edge_to_coordinate(l1, l2, e)
    lines.append(cdn)

lc = LineCollection(lines, linewidth = 6,color = "gray", label="Erasure error")
ax.add_collection(lc)

zlines = []
for z in Z_errors_on_qubits:
    cdn = edge_to_coordinate(l1, l2, z)
    zlines.append(cdn)

lc = LineCollection(zlines, linewidth = 2,color = "blue", label="Z error")
ax.add_collection(lc)
# faces, X stab
for i in faces:
    p = pat.Polygon(xy = [i[0], i[2], i[3], i[1]], ec = "gray", color = "lightgray")
    ax.add_patch(p)

first_time_to_show = True

for x_synd in X_syndromes:
    cdn = face_to_coordinate(l2, x_synd)
    if first_time_to_show:
            p = pat.Polygon(xy = [cdn[0], cdn[2], cdn[3], cdn[1]], ec = "red", color = "lightpink", label="X stabilizer with -1")
            first_time_to_show = False
    else:
        p = pat.Polygon(xy = [cdn[0], cdn[2], cdn[3], cdn[1]], ec = "red", color = "lightpink")
    ax.add_patch(p)

plt.legend(loc='upper left')
plt.xlim(-1,l2 + 1)
plt.ylim(-1,l1 + 2)
plt.show()
```
